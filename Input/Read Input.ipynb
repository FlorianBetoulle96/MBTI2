```python
import pandas as pd
#from ast import literal_eval
import re

from collections import Counter
```

```python
# Read the CSV file into a DataFrame
bag_raw = pd.read_csv('bag_raw.csv')

# Filter lines where the speaker is 'Michael' and extract the 'line_token'
bag_one_speaker = bag_raw.loc[bag_raw['speaker'] == 'Roy', 'line_token'].values[0]

# Extract the content inside Counter() using regex
counter_content = re.search(r"Counter\({(.+)}\)", bag_one_speaker).group(1)

# Split by comma and colon to get key-value pairs
items = [item.strip().split(": ") for item in counter_content.split(",")]

# Create a DataFrame from the parsed items
df = pd.DataFrame(items, columns=['letter', 'occurrence'])

# Clean up the 'letter' column (remove quotes and braces)
df['letter'] = df['letter'].str.strip("'").str.strip()

# Convert 'occurrence' column to integers
df['occurrence'] = df['occurrence'].astype(int)
df_short = df[df['occurrence'] > 0]

pd.set_option('display.max_rows', None)

# Print the resulting DataFrame
df_short
```
